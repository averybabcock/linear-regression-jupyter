Linear Model for R

This imports the necessary tools for creating the plot and statistics

```r
library(ggplot2)
```

Set the file and the columns

```r
filename = 'regression_data.csv'
x_col = 'YearsExperience'
y_col = 'Salary'
```

Loads csv

```r
data <- read.csv(filename)
```

Fit the plot and perform linear regression to get line of best fit, the intercepts from the data file, the p value which can be used to test the significance of the data, standard error, and then the equation for the plot, which is based on the scatter data.

The last portion calculates mean squared error between the predictions and the actual values. This quantifies how close the model predicts the actual data.It emphasizes large errors more than smaller ones due to squaring.The formula for MSE is MSE = (1/n)*Σ (yᵢ - ŷᵢ)² and the units are squared units of y. The smaller the MSE, the better fit the model is.

```r
formula <- as.formula(paste(y_col, "~", x_col))
model <- lm(formula, data = data)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r <- cor(data[[x_col]], data[[y_col]])
pred <- predict(model)
mse <- mean((data[[y_col]] - pred)^2)
```

Prepare the annotation text for the plot

```r
equation_text <- paste0(
  "Equation: y = ", round(slope, 2), "x + ", round(intercept, 2), "\n",
  "Correlation (r): ", round(r, 2), "\n",
  "MSE: ", round(mse, 2)
)
```

Make edits to plot color, axis, title, and annotations. Then save as .png and then print

```r
plot <- ggplot(data, aes_string(x = x_col, y = y_col)) +
  geom_point(color = "red") +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  ggtitle(paste(y_col, "vs", x_col)) +
  xlab(x_col) +
  ylab(y_col) +
  annotate(
    "text",
    x = min(data[[x_col]]),
    y = max(data[[y_col]]),
    label = equation_text,
    hjust = 0,
    vjust = 1,
    size = 5,
    color = "black"
  )

ggsave("linear_regression_r_output.png", plot)
print(plot)
```
